```python
val = linspace(10, 20, 5)
for x in val:
    print(x, end=" , ")
    
# Range = 20 - 10 = 10
# Gaps = 5 - 1 = 4
# Gap size = 10 / 4 = 2.5
```

```python
val = arange(10, 20, 2)
for x in val:
    print(x, end=" , ")
```

```python
val = logspace(11, 20, 2)     # logspace(start, stop, count)
for x in val:
    print(x, end=" ")
    
# logspace(11, 20, 2) ka matlab
# Start power = 11 ‚Üí 10¬π¬π
# Stop power = 20 ‚Üí 10¬≤‚Å∞
# Total values = 2

# üîπ Generated Values
# First value = 10¬π¬π
# Second value = 10¬≤‚Å∞
```
