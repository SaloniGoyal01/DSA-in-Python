```python
from array import *
val = array('i', [10, 20, 30, 40, 50])
copyArray = array(val.typecode, (x*3 for x in val))
copyArray.pop(3)
for i in range(0, len(val)):
    print(val[i], end=" ")
for i in range(0, len(copyArray)):
    print("\n")
    print(copyArray[i], end=",")
```

```python
val = array('i', [10, 20, 30, 40, 50, 60, 70, 80])
x = val[2:-3]    # start index 2 and last ke 3 ht jayenge -3 ke karan
for i in range(0, len(x)):
    print(x[i], end=" ")
```
