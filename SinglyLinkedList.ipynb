```python
# ---------------- NODE CLASS ----------------
# Node = Linked List ka ek dabba
# Isme 2 cheeze hoti hain:
# 1. info → data
# 2. next → next node ka address

class Node:
    def __init__(self, info):
        self.info = info      # data store hota hai
        self.next = None     # next node ka address (starting me None)


# ---------------- SINGLY LINKED LIST CLASS ----------------
class SinglyLL:
    def __init__(self):
        self.head = None     # head = first node (starting point)
 

    # ---------- INSERT AT END ----------
    # List ke last me new node add karta hai
    def insertAtEnd(self, value):
        new_node = Node(value)    # new node banao

        # agar list empty hai
        if self.head is None:
            self.head = new_node
            return

        # last node tak pahunchne ke liye traversal
        temp = self.head
        while temp.next is not None:
            temp = temp.next

        # last node ke next me new node jod do
        temp.next = new_node


    # ---------- INSERT AT BEGINNING ----------
    # List ke start me new node add karta hai
    def insertAtBeginning(self, value):
        new_node = Node(value)    # new node

        new_node.next = self.head   # new ka next = purana head
        self.head = new_node        # head update


    # ---------- INSERT AFTER A GIVEN VALUE ----------
    # Kisi value ke baad new node add karta hai
    def insertAtMid(self, newValue, afterValue):
        temp = self.head

        # poori list traverse karo
        while temp is not None:
            if temp.info == afterValue:
                new_node = Node(newValue)

                new_node.next = temp.next
                temp.next = new_node
                return
            temp = temp.next


    # ---------- DELETE A NODE ----------
    # Given value wali node delete karta hai
    def deleteLL(self, value):

        # case 1: agar delete karni wali node head hai
        if self.head is not None and self.head.info == value:
            self.head = self.head.next
            return

        temp = self.head
        prev = None

        # list traverse karo
        while temp is not None:
            if temp.info == value:
                prev.next = temp.next
                return
            prev = temp
            temp = temp.next


    # ---------- PRINT LINKED LIST ----------
    # Linked list ko print karta hai
    def printLL(self):
        temp = self.head
        while temp is not None:
            print(temp.info)
            temp = temp.next
        print("None")

obj = SinglyLL()

obj.insertAtEnd(10)
obj.insertAtEnd(20)
obj.insertAtEnd(30)
obj.insertAtBeginning(5)
obj.insertAtMid(40, 20)
obj.deleteLL(20)
obj.printLL()
```
